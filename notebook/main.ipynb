```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import logging

# Suppress webdriver_manager logs
logging.getLogger("WDM").setLevel(logging.NOTSET)

CHROMEDRIVER_PATH = "/home/neutrino/.wdm/drivers/chromedriver/linux64/133.0.6943.141/chromedriver-linux64/chromedriver"

def setup_driver():
    options = Options()
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    options.add_argument("--window-size=1920x1080")
    options.add_argument("--headless")  # Run in background
    return webdriver.Chrome(service=Service(CHROMEDRIVER_PATH), options=options)
```

```python
def get_google_trends():
    driver = setup_driver()
    url = "https://trends.google.com/trending?geo=PK"
    driver.get(url)

    time.sleep(5)  # Adjust sleep time as necessary

    trends = driver.find_elements(By.CLASS_NAME, "mZ3RIc")
    
    trend_list = []
    for trend in trends:
        trend_text = trend.text
        if trend_text:
            trend_list.append(trend_text)

    driver.quit()
    
    return trend_list

get_google_trends()
```

```python
def get_youtube_trends():
    driver = setup_driver()
    driver.get("https://www.youtube.com/feed/trending")
    time.sleep(5)  # Initial wait for page to load

    driver.execute_script("window.scrollTo(0, 2000)")
    time.sleep(2)

    video_elements = driver.find_elements(By.CSS_SELECTOR, "ytd-video-renderer, ytd-grid-video-renderer")

    data = {
        "title": [],
        "channel": [],
        "metadata": []
    }

    for index, video in enumerate(video_elements, start=1):
        try:
            title_element = video.find_element(By.CSS_SELECTOR, "#video-title")
            title = title_element.text or title_element.get_attribute("title")
            
            channel = video.find_element(By.CSS_SELECTOR, "yt-formatted-string.ytd-channel-name a").text
            
            metadata = video.find_element(By.CSS_SELECTOR, "span.style-scope.ytd-video-meta-block").text
            
            data["title"].append(title)
            data["channel"].append(channel)
            data["metadata"].append(metadata)
        except Exception as e:
            print(f"Error processing video {index}: {str(e)}")
    driver.quit()
    
    return data

get_youtube_trends()
```

```python
from gnews import GNews

def get_news_trends():
    
    news_data = {
        "title": [],
        "description": [],
        "published date": [],
        "publisher": []
    }
    
    news_client = GNews(language='en', period='24h')
    trending_news = news_client.get_top_news()
    
    for trend in trending_news:
        news_data["title"].append(trend['title'])
        news_data["description"].append(trend['description'])
        news_data["published date"].append(trend['published date'])
        news_data["publisher"].append(trend['publisher']['title'])

    return news_data

get_news_trends()
```

```python
import concurrent.futures

def analyze_trends_parallel():
    # Create a thread pool with 3 workers (one for each API)
    with concurrent.futures.ThreadPoolExecutor(max_workers=3) as executor:
        # Submit all tasks to the executor
        google_future = executor.submit(get_google_trends)
        youtube_future = executor.submit(get_youtube_trends)
        news_future = executor.submit(get_news_trends)
        
        # Wait for all tasks to complete and get results
        google_trends = google_future.result()
        youtube_trends = youtube_future.result()
        news_trends = news_future.result()
        
        return google_trends, youtube_trends, news_trends

    # # Print results (same as before)
    # print("Google Trends:")
    # for trend in google_trends:
    #     print(trend)

    # print("\nYouTube Trends:")
    # for title, channel, metadata in zip(youtube_trends["title"], youtube_trends["channel"], youtube_trends["metadata"]):
    #     print(f"Title: {title}, Channel: {channel}, Metadata: {metadata}")

    # print("\nNews Trends:")
    # for title, description, published_date, publisher in zip(news_trends["title"], news_trends["description"], news_trends["published date"], news_trends["publisher"]):
    #     print(f"Title: {title}, Description: {description}, Published Date: {published_date}, Publisher: {publisher}")

# Run the parallel version
analyze_trends_parallel()
```

```python
from langchain_ollama import ChatOllama 
from langchain.prompts import PromptTemplate
from langchain_core.runnables import RunnablePassthrough
```

```python
llm = ChatOllama(model="llama3.2:3b", temperature=0.1, max_tokens=1000)

prompt = PromptTemplate(
    input_variables=["trends", "theme"],
    template="""
You are a creative writer tasked with crafting a short story in the style of a {theme} tale.

Below are the trending topics from the past 24 hours. Your job is to subtly weave their essence, themes, or events into a fictional narrative. Do **not** list them as-is — instead, use them as **inspiration** for characters, situations, or plot twists.

Trending Topics:
{trends}

Guidelines:
- Use a natural, engaging storytelling voice.
- The tone must match the selected theme: {theme}.
- Be imaginative — you can use metaphors, symbolism, or satire.
- Avoid making it feel like a news recap.
- Keep it concise, yet emotionally or humorously impactful.

Begin the story below:
"""
)

chain = prompt | llm
```

```python
def generate_trend_story(trends, theme="comedy"):
    # Run through LangChain with Ollama
    story = chain.invoke({"trends": trends, "theme": theme})
    return story.content
```
